### gender and salary

```python
male_sal = df[df['Gender']=='Male']['Salary_INR']
```

```python
male_sal
```

```python
female_sal = df[df['Gender']=='Female']['Salary_INR']
```

```python
female_sal
```

```python
from scipy import stats
import numpy as np
```

```python
from scipy.stats import ttest_ind


t_test,p_value =ttest_ind(male_sal,female_sal,equal_var=False)
```

```python
if p_value < 0.05:
    print("Reject H₀ → Attrition depends on Gender (significant relationship")
else:
    print("Fail to Reject H₀ → Attrition independent of Gender")
```

### gender and attrition_status

```python
contingency = pd.crosstab(df['Gender'], df['Attrition_Status'])

chi2_stat, p_val, dof, expected = stats.chi2_contingency(contingency)

print(f"Chi2 Statistic: {chi2_stat:.4f}, P-value: {p_val:.4f}")

if p_val < 0.05:
    print("Reject H₀ → Attrition depends on Gender (significant relationship).\n")
else:
    print("Fail to Reject H₀ → Attrition independent of Gender.\n")
```
